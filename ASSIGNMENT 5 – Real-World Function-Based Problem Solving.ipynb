```python
def analyze(l):
    t=0
    for i in l:
        t+=i
    if t>=1000:
        s="Viral Post"
    else:
        s="Normal Engagement"
    return t,s
a=[250,300,150,200,150]
t,s=analyze(a)
print("Total Likes:",t)
print("Post Status:",s)
```

```python
def check(s):
    if s<10:
        r="Low Stock Alert"
    else:
        r="Stock Sufficient"
    return s,r
a=6
s,r=check(a)
print("Medicine Stock:",s)
print("Status:",r)
```

```python
def check(a,r):
    t=0
    for i in a:
        t+=i
    avg=t/len(a)
    if avg>=r:
        s="Adequate Rainfall"
    else:
        s="Inadequate Rainfall"
    return avg,s
d=[70,75,80,65,70]
req=70
avg,s=check(d,req)
print("Average Rainfall:",int(avg))
print("Rainfall Status:",s)
```

```python
def check(a):
    s=set(a)
    if len(s)<len(a):
        r="Yes"
    else:
        r="No"
    return r
u=["ram","sita","john","ram","alex"]
r=check(u)
print("Duplicate Accounts Found:",r)
```

```python
def check(a,t):
    p=[]
    for i in a:
        if i>t:
            p.append(i)
    return p
a=[1500,2500,1800,3200,2100]
t=2000
p=check(a,t)
print("Premium Crops:",p)
```

```python
def check(a,l):
    if a<=l:
        s="Approved"
    else:
        s="Rejected"
    return a,s
a=60000
l=50000
a,s=check(a,l)
print("Transaction Amount:",a)
print("Transaction Status:",s)
```

```python
def check(a):
    p=0
    for i in a:
        if i=="P":
            p+=1
    per=(p/len(a))*100
    if per>=75:
        s="Eligible"
    else:
        s="Not Eligible"
    return per,s
a=["P","P","A","P","P"]
per,s=check(a)
print("Attendance Percentage:",per)
print("Exam Eligibility:",s)
```

```python
def check(u):
    if u<=100:
        b=u*3
    elif u<=200:
        b=(100*3)+((u-100)*5)
    else:
        b=(100*3)+(100*5)+((u-200)*7)
    if b<500:
        s="Low Usage"
    elif b<=1500:
        s="Moderate Usage"
    else:
        s="High Usage"
    return b,s
u=250
b,s=check(u)
print("Total Bill:",b)
print("Usage Status:",s)
```

```python
def lower_text(text):
    return text.lower()
def remove_links(text):
    return re.sub(r'http\S+|www\S+', '', text)
def remove_emoji(text):
    return emoji.replace_emoji(text, replace='')
def remove_symbols(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
def tokenize_text(text):
    return word_tokenize(text)
def remove_stop_words(tokens):
    sw = set(stopwords.words('english'))
    return [i for i in tokens if i not in sw]
def stem_words(tokens):
    ps = PorterStemmer()
    return [ps.stem(i) for i in tokens]
def lemma_words(tokens):
    lm = WordNetLemmatizer()
    return [lm.lemmatize(i) for i in tokens]
```

```python
def process_text(text):
    text = lower_text(text)
    text = remove_links(text)
    text = remove_emoji(text)
    text = remove_symbols(text)
    tokens = tokenize_text(text)
    tokens = remove_stop_words(tokens)
    stem = stem_words(tokens)
    lemma = lemma_words(tokens)
    return tokens, stem, lemma
```

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    if total == 0:
        avg_len = 0
    else:
        avg_len = sum(len(w) for w in tokens) / total
    return total, unique, avg_len
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for t in text_list:
        tokens, clean = preprocess_text(t)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    if text == "":
        return [], ""
    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]
    clean_text = ' '.join(tokens)
    return tokens, clean_text
```

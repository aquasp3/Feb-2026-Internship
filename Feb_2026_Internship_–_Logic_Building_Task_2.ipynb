```python
# Unique Words in a Sentence

a="python is easy and python is powerful"
b=a.split()
unique=set(b)
print("the count of the unique words:",len(unique))
print("the unique words are :",unique)
```

```python
#Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
a=max(employees,key=employees.get)
print("highest salary :",a,":",employees[a])
```

```python
#Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
max=numbers[0]
min=numbers[0]
for i in numbers:
    if i>max:
        max=i
    if i<min:
        min=i
print("max:",max)
print("min:",min)
```

```python
#Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
  if i>1000:
    count+=1
print("products above 1000:",count)
```

```python
#Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
  if i =="P":
    count+=1
mean=count/len(attendance)*100
print("attendance percentage :",mean)
```

```python
#Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]
unique=set(phone_numbers)
print("unique phone numbers:",unique)
```

```python
#Count Character Frequency

a="banana"
char={}
for i in a:

  if i in char:
    char[i]+=1
  else:
    char[i]=1
print(char)
```

```python
#Calculate Average Marks

marks=[56,78,85,90,94]
sum=0
for i in marks:
  sum=sum+i
mean=sum/len(marks)
print("average marks:",mean)
```

```python
def validate_recharge(amount):
    valid_plans = (199, 299, 399, 599)
    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    if amount not in valid_plans:
        print("Invalid plan. Choose from 199, 299, 399, 599")
        return False
    print("Recharge successful")
    return True
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        if validate_recharge(amount):
            break
        print("Try again\n")
    except ValueError:
        print("Enter a valid number\n")
```

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(product + " : Reorder Alert")
        else:
            print(product + " : Stock OK")
inventory = {
    "Rice": 40,
    "Sugar": 12,
    "Oil": 8,
    "Soap": 25,
    "Salt": 14
}
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result

student_marks = [60, 45, 70, 55, 40]
avg, status = process_result(student_marks)
print("Average:", avg)
print("Result:", status)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)

    return fare
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")
        total_fare = calculate_fare(distance, peak)
        print("Total Fare: ₹", round(total_fare, 2))
        again = input("Do you want to calculate again? (yes/no): ")
        if again.lower() != "yes":
            break
    except ValueError:
        print("Enter valid input\n")
```

```python
def check_eligibility(attendance):
    present_count = 0
    for day in attendance:
        if day == "P":
            present_count += 1
    percentage = (present_count / len(attendance)) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return percentage, status
records = ["P", "A", "P", "P", "A", "P", "P", "P"]
percent, result = check_eligibility(records)
print("Attendance Percentage:", percent)
print("Status:", result)
```

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```

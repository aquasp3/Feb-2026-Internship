```python
#User Login Check
username="admin"
password="1234"
a=input("enter username :")
b=input("enter password :")
if a==username and b==password:
    print("Login Successful")
else:
    print("Login Failed")
```

```python
#pass or fail
marks=[45,78,90,33,60]
passed=[]
fail=[]
for i in marks:
    if i>=50:
        passed.append(i)
    else:
        fail.append(i)
print("total number of passed students",len(passed))
print("total number of failed students",len(fail))
```

```python
#simple data cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Cleaned Names:", cleaned_names)
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    print(f"message: {message}, Length: {len(message)}")
    if len(message)>10:
      print("message longer than 10 characters")
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
  if i=="ERROR":
    count+=1

print("total no of errors :",count)
```

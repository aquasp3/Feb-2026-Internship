```python
#Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
highest_score=0
for name,score in employees.items():
  if score>highest_score:
    highest_score=score
top_scorer=[]
for name,score in employees.items():
  if score==highest_score:
    top_scorer.append(name)
print("Top Performers Eligible for Bonus:",",".join(top_scorer),"(score:",highest_score,")")


```

```python
# Search Query Keyword Analysis
a="Buy mobile phone buy phone online"
b=a.lower()
c=b.split()
char={}
for i in c:
  if i in char:
    char[i]+=1
  else:
    char[i]=1
for i in char:
  if char[i]>1:
    print(i,":",char[i])
```

```python
#Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
char={}
for i in sensor_readings:
  if i%2==0:
    a=sensor_readings.index(i)
    char[a]=i
b=[]
for i,j in char.items():
  b.append((i,j))
print(b)
```

```python
#Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain={}
for i in emails:
  a=i.split("@")[1]
  if a in domain:
    domain[a]+=1
  else:
    domain[a]=1
total=len(emails)
for i in domain:
  print(i,":",(domain[i]/total)*100,"%")
```

```python
#Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
total=len(sales)
mean=sum(sales)/total

for index,sale in enumerate(sales):
  if sale>(mean+(mean*0.3)):
    print("Day",index+1,":",sale)



```

```python
#Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
char={}
for i in user_ids:
  if i in char:
    char[i]+=1
  else:
    char[i]=1
for i in char:
  if char[i]>1:
    print(i,"->",char[i],"times")
```
